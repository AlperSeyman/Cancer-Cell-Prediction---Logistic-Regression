```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data.info()
```

```python
data.drop(["Unnamed: 32","id"], axis=1, inplace=True)
```

```python
data.head()
```

```python
data["diagnosis"] = [ 1 if i=="M" else 0 for i in data["diagnosis"]]
```

```python
data.head()
```

```python
y = data["diagnosis"].values
x = data.drop("diagnosis",axis=1)
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x = scaler.fit_transform(x)
x = pd.DataFrame(x, columns=data.drop("diagnosis",axis=1).columns)
x.head()
```

```python
# Train and Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()

logistic_model = logistic_regression.fit(x_train, y_train)

# Logistic Regression Model Score
logistic_model.score(x_test, y_test)
```

```python
# Logistic Regression Predicition
y_pred = logistic_regression.predict(x_test)

# Actual and Prediction Result
result = pd.DataFrame({"Actual":y_test,"Prediction":y_pred})
result.head()
```
